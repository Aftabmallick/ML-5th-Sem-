```python
import numpy as np
```

```python
x=np.arange(0,100,1)
```

```python
x
```

```python
y=x*2
```

```python
y
```

```python
z=x**2
```

```python
z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'r--')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title("X Y graph")
```

```python
figure=plt.figure()
ax1=figure.add_axes([0,0,1,1])
ax2=figure.add_axes([0.1,0.5,.4,.4])
ax1.plot(x,y,'r--')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('ax1')

ax2.plot(x,y,'g--')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_title('ax2')
```

```python
figure=plt.figure()
ax1=figure.add_axes([0,0,1,1])
ax2=figure.add_axes([0.1,0.5,.4,.4])
ax1.plot(x,z,'r--')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('ax1')

ax2.plot(x,y,'g--')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_title('Zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b--',lw=3)
axes[1].plot(x,z,'r--',lw=5)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,5))
axes[0].plot(x,y,'b--',lw=3)
axes[1].plot(x,z,'r--',lw=5)
```
